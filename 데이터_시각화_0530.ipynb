```python
import pandas as pd
import seaborn as sns


csv_file_path = '/content/body1.csv'  # Replace with the actual path to your CSV file
body_df = pd.read_csv('/content/body1.csv')
body_df.head()
```

```python
body_df.plot(kind = 'hist', y = 'Height', bins = 15) #bins = 15 해서 막대기가 15개 생김.
```

```python
sns.histplot(body_df['Height'], bins = 15)
```

```python
#박스플롯
body_df.plot(kind = 'box', y = 'Height')
```

```python
sns.violinplot(y = body_df['Height']) #분포 전체를 보여준다.
```

```python
#키와 몸무게의 연관성 (산점도)

body_df.plot(kind = 'scatter', x = 'Height', y = 'Weight')
```

```python
#좀더 멋잇게 보기(등고선형)

sns.kdeplot(x = body_df['Height'], y=  body_df['Weight'])
```

```python
#몸무게 예측(회귀선)

sns.lmplot(data = body_df, x = 'Height', y = 'Weight')
```

```python
import pandas as pd
import seaborn as sns


csv_file_path = '/content/laptops1.csv'  # Replace with the actual path to your CSV file
laptop_df = pd.read_csv('/content/laptops1.csv')
laptop_df.head()
```

```python
laptop_df['os'].unique() #판다스 시리즈.
```

```python
sns.catplot(data = laptop_df, x ='os', y = 'price', kind = 'box')
```

## 각 운영체제의 갯수를 정확히 알 수 없어 정확한지 알 수 없다.
### 그럴때는 kind = 'strip' 써서 표현.

```python
sns.catplot(data = laptop_df, x = 'os', y = 'price', kind = 'strip', hue='processor_brand', palette='Set1')
```

```python
laptop_df['processor_brand'].unique()
```

# swarm 플롯
### 뭉쳐진 점들을 펼쳐서 보기쉽게

```python
sns.catplot(data = laptop_df, x = 'os', y = 'price', kind = 'swarm', hue='processor_brand', palette='Set1')
```
