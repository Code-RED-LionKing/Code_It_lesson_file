# 상관 관계 분석(Correlation Analysis)

```python
csv_file_path = '/content/young_survey.csv'  # Replace with the actual path to your CSV file
music_df= pd.read_csv('/content/young_survey.csv')
music_df
```

```python
#
music = music_df.iloc[:, :19]
music.head()
```

```python
sns.heatmap(music.corr())
#연할수록 연관도가 높다
```

```python
music.corr()

```

- corr() 메서드는 데이터프레임의 열들 간의 상관관계를 계산하는 데 사용됩니다. 상관관계는 두 변수 간의 선형적인 관계를 측정하는데 도움이 됩니다. 이 메서드는 주로 수치형 데이터에 대해 사용되며, 각 변수 간의 상관계수를 계산하여 변수 간의 관계를 평가할 수 있습니다.

- corr() 메서드는 기본적으로 피어슨 상관계수(Pearson correlation coefficient)를 계산합니다. 상관계수는 -1부터 1까지의 값을 가지며, -1은 완전히 음의 선형 관계, 1은 완전히 양의 선형 관계를 의미하며, 0은 선형 관계가 없음을 의미합니다.

- 일반적으로 corr() 메서드는 다음과 같은 경우에 사용됩니다:

- 변수 간의 상관관계를 확인하고 분석하는 경우
데이터프레임의 열들 간의 상관관계를 시각화하려는 경우 (heatmap 등)
변수 선택이나 다중공선성(multicollinearity) 검사와 같은 분석 작업에 사용하는 경우
예를 들어, music 데이터프레임의 열들 간의 상관관계를 확인하고 싶다면, music.corr()을 호출하여 상관계수 행렬을 얻을 수 있습니다. 이를 통해 각 변수 간의 관계를 이해하고 해석할 수 있습니다.

```python
music_df.corr()['Age'].sort_values(ascending=False)
```
